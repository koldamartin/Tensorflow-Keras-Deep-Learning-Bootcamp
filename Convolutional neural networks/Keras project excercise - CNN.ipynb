# **Deep learning for Image Classification Excercise**

## **The Challenge**
Your task is to build an image classifier with Keras and Convolutional Neural Networks for the Fashion MNIST dataset. This data set includes 10 labels of different clothing types with 28 by 28 grayscale images. There is a training set of 60,000 images and 10,000 test images.

**Label	Description**
- 0	    T-shirt/top
- 1	    Trouser
- 2	    Pullover
- 3	    Dress
- 4	    Coat
- 5	    Sandal
- 6	    Shirt
- 7	    Sneaker
- 8	    Bag
- 9	    Ankle boot**

## **Import the data**

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from keras.datasets import fashion_mnist
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data();
```

##  Preprocessing the data

```python
x_train.shape
```

```python
x_test.shape
```

```python
plt.imshow(x_train[0],cmap='Greys')
```

```python
x_train[0]
```

```python
# Check the color range
x_train.max()
```

**Normalizing the data**

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_train.max()
```

**The labels**

```python
y_train.shape
```

```python
y_test.shape
```

```python
print(f"The label for image [0] is {y_train[0]}")
# number 9 is Ankle boot, that is correct
```

```python
# One hot encoding of the 10 labels
from tensorflow.keras.utils import to_categorical
```

```python
y_train_cat = to_categorical(y_train, num_classes=10)
```

```python
y_test_cat = to_categorical(y_test, num_classes=10)
```

```python
y_train_cat[0]
```

**Reshaping the data** \
CNN needs an extra dimension for each color channel, 1 for greyscale and 3 for RGB

```python
x_train = x_train.reshape(60000, 28, 28, 1)
x_test = x_test.reshape(10000, 28, 28, 1)
```

```python
print(f"train shape: {x_train.shape}")
print(f"test shape: {x_test.shape}")
```

## Modelling the CNN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten
```

```python
model = Sequential()

# Convolutional layer
model.add(Conv2D(filters=12, kernel_size = (4,4), strides=2, activation='relu'))
#Pooling layer
model.add(MaxPool2D(pool_size=(2,2)))
# Flattern the images from 28x28 to 784 before final layer
model.add(Flatten())
# Hidden dense layer
model.add(Dense(60, activation='relu'))
# Last output layer is the classifier for 10 classes
model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=2)
```

**Training the model**

```python
model.fit(x_train, y_train_cat, epochs=10, validation_data=(x_test, y_test_cat), callbacks=[early_stop])
```

**Evaluating the model**

```python
pd.DataFrame(model.history.history)[['accuracy', 'val_accuracy']].plot()
```

```python
pd.DataFrame(model.history.history)[['loss', 'val_loss']].plot()
```

```python
print(model.metrics_names)
print(model.evaluate(x_test, y_test_cat,verbose=0))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_pred = model.predict(x_test)
y_pred[0]
```

```python
# because model.predict_classes is deprecated, i will use np.argmax to assign 1 to highest number
y_pred = np.argmax(y_pred, axis=1)
```

```python
y_pred[0]
```

```python
y_test[0]
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True,cmap='viridis')
```

# Predict a given image

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
def tell_the_image(image_nr):
    plt.imshow(x_test[image_nr])
    images = x_test[image_nr].reshape(1,28,28,1)
    predicted_image = model.predict(images)
    predicted_image = np.argmax(predicted_image, axis=1)
    for x in range(len(class_names)):
        if predicted_image == x:
            print(f'This is a picture of {class_names[x]}')
```
